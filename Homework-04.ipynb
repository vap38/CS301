# Problem 1
Write a function called get_percent_nans which takes the following input parameters: ● df (DataFrame) - a DataFrame object ● column_name (str) - a column name

```python
import pandas as pd
import numpy as np
```

```python
def get_percent_nans(df,column_name):
    f1 = pd.isnull(df[column_name])
    empty = len(df[f1])
    length = len(df[column_name])
    percent= (empty/length)*100
    return round(percent,2)
```

```python
my_df = pd.read_csv("2019_nCoV_20200121_20200206.csv")
get_percent_nans(my_df,"Death")
```

# Problem 3

# Problem 4
According to the data found in the “2019_nCoV_20200121_20200206.csv” file, which
Country/Region has the second highest average number of recovered cases of 2019 nCoV?

```python
recovered = my_df.groupby('Country/Region',as_index=False)['Recovered'].mean()
recovered_sorted = recovered.sort_values("Recovered",ascending =False)
print(recovered_sorted.iloc[1])
```

```python
def get_continent(country):
    d = {"Asia": ["Mainland China","India","Russia","Singapore","Thailand","Singapore","Hong Kong","South Korea","Malaysia","Japan","Vietnam","Taiwan","United Arab Emirates","Phillipines","Nepal","Cambodia","Sri Lanka","Macau"],"Africa": ["Ivory Coast"],"Australia":["Australia"],
                 "Antarctica":[],"Europe":["Spain","Italy","UK","Germany""Belgium","Sweden","Finland"],"North America":["United States","Canada"],"South America":["Colombia","Mexico","Brazil"]}
    for key in d.keys():
        
        if(country in d[key]):
            return key
        
```

```python
get_continent("United States")
```
